# Problema de min-max

Resolución grafica del problema

$\mathrm{minimizar}_{x\in R^2} \max \{ \|x-y_1\|^2,\|x-y_2\|^2,\ldots,\|x-y_p\|^2  \}$
donde $y_i,\;i=1,2,\ldots,p$ son vectores de $R^2$ fijos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generar n puntos aleatorios dentro del cuadrado de -10 a 10
# np.random.seed(42)  # Para obtener resultados reproducibles
n = 4
puntos = np.random.uniform(-10, 10, size=(n, 2))

# Función para calcular la distancia máxima a los puntos
def distancia_maxima(p, puntos):
    return max(np.linalg.norm(p - punto) for punto in puntos)

# Crear una malla de puntos en el plano entre -10 y 10
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)

# Calcular la distancia máxima para cada punto en la malla
Z = np.zeros_like(X)
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        p = np.array([X[i, j], Y[i, j]])
        Z[i, j] = distancia_maxima(p, puntos)

# Graficar las curvas de nivel
plt.figure(figsize=(8, 6))
cp = plt.contour(X, Y, Z, 100, cmap='viridis')  # 20 niveles de contorno
plt.colorbar(cp)  # Barra de color para mostrar los valores de la distancia
plt.scatter(puntos[:, 0], puntos[:, 1], color='red')  # Puntos generados
plt.title('Curvas de nivel de la distancia máxima a los puntos aleatorios')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```
